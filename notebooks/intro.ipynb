```python
import pymupdf

def extract_text_from_pdf(pdf_path):
    pdf_document = pymupdf.open(pdf_path)
    pages = []
    for page in pdf_document:
        pages.append(page.get_text())
    pdf_document.close()
    return pages
```

```python
extracted_text = extract_text_from_pdf("../data/Kenya_Health_Facility_Census_Report_September_2023.pdf")
```

```python
from langchain.vectorstores.chroma import Chroma
from langchain_core.prompts import ChatPromptTemplate
from langchain_google_genai import GoogleGenerativeAIEmbeddings
import os 
import getpass

if "GOOGLE_API_KEY" not in os.environ:
    os.environ["GOOGLE_API_KEY"] = getpass.getpass("Enter your API key: ")

embeddings = GoogleGenerativeAIEmbeddings(model="models/text-embeeding-004")
```

```python
from langchain.docstore.document import Document

docs = []

# for i, page in enumerate(extracted_text):
#     docs.append(Document(content=page, title=f"Page {i+1}"))

for p in extracted_text:
    doc = Document(page_content=p, metadata={"source": "local"})
    docs.append(doc)

# docs
db = Chroma.from_documents(
    docs,
    embeddings
)
```
